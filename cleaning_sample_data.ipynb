# Function Definitions

```python
import pandas as pd 
import numpy as np
```

```python
def load_date(file_path):
    df = pd.read_csv(file_path, encoding='utf-8')
    print("Shape of dataframe:", df.shape)
    return df
```

```python
def add_column_index(df):
    df.insert(0, "index", range(1, len(df) + 1))
    return df
```

```python
def Clean_Binaries(df, col, mapping):
    print(f"{col} unique values:\n", df[col].unique())
    
    # Convert to binary values and clean it:
    # Words that are not in the mapping, will be converted to NaN
    df[col] = df[col].str.strip()
    df[col] = df[col].map(mapping)
    print(f"{col} unique values after process:\n", df[col].unique())

    return df
```

```python
def check_null_dup(df, col):
       
    if df1[df1[col].isna()].shape[0] != 0:
        print(f"Null {col} Found!")
    else:
        print(f"No Null {col}")
        
    if len(df[col].unique()) != df.shape[0]:
        if df[df.duplicated(col, keep=False)].shape[0] != 0:
            print(f"Duplicated {col}!")
            print(df[df.duplicated(col, keep=False)][["index", col]])
```

```python
def user_regex(df, col, reg):
    print(f"{col} unique values:\n", df[col].unique())

    # Extract matched ones from strings and make all others, nan
    df[col] = (
        df[col]
        .astype(str)
        .str.extract(reg, expand=False)  
        .astype(float)
    )
    print(f"{col} unique values after process:\n", df[col].unique())
    return df
```

```python
def Clean_Numerics(df, col, mapping):
    print(f"{col} unique values:\n", df[col].unique())

    # 1. Convert to valid values:
    # Words that are not in the mapping, will be unchanged
    df[col] = df[col].replace(mapping)

    # 2. Convert numeric-looking values to number, errors become NaN
    df[col] = pd.to_numeric(df[col], errors="coerce")
    
    print(f"{col} unique values after process:\n", df[col].unique())

    return df
```

```python
def not_numbers_to_nan(df, col):
    print(f"{col} unique values:\n", df[col].unique())

    df[col] = pd.to_numeric(
        df[col].astype(str).str.strip(),
        errors="coerce"
    )
    
    print(f"{col} unique values after process:\n", df[col].unique())
    return df
```

```python
def clean_length_cols(df, col):
    print(f"{col} unique values:\n", df[col].unique())

    s = df[col].astype(str)
    
    mm_vals = s.str.extract(r'(\d+(?:[.,]\d+)?)\s*mm', expand=True)[0]  # integer OR decimal number (',' or '.' as decimal separator) + zero or more spaces + "mm"
    
    num_vals = s.str.extract(r'^\s*(\d+(?:[.,]\d+)?)\s*$', expand=True)[0] # string must contain only a number (optionally with spaces around it)
    
    # Combine the extracted values, filling NaNs from mm_vals with num_vals
    combined_vals = mm_vals.fillna(num_vals)
    
    # Replace comma with dot for decimal conversion and convert to numeric
    df[col] = pd.to_numeric(
        combined_vals.str.replace(',', '.', regex=False),
        errors="coerce"
    )
    
    print(f"{col} unique values after processing:\n", df[col].unique())

    return df
```
